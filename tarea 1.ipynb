```python
L=1000


N=10
x=np.linspace(0,L,N)
dx=L/N-1
#condiciones
t=0
Pini=4000
P=np.ones(N)*Pini
P0=0
P[0]=P0
PL=0
P[-1]=PL


#solu
dt=10
tfin=180
Pdt=P.copy()
Psol=[P]
tsol=[t]

while t<tfin:
  for i in range(N):
    if i==0:
      Pdt[i]=P0
    elif i==N-1:
       Pdt[i]=PL
    else:
        Pdt[i]=(((P[i-1]-2*P[i]+P[i+1])/(dx**2))*dt)+P[i]
  P=Pdt.copy()
  t=t+dt
  print(P)

```
